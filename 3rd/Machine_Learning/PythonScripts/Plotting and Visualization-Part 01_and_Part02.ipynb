```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,10)
y = 2 * x
```

```python
plt.plot(x,y) 
```

# Second plot on the top of first plot

```python
plt.plot(x,y)
plt.plot(y,x)
```

# Lock the plot by plt.show()

```python
plt.plot(x,y)
plt.show()
plt.plot(y,x)
```

# Change the color of the plot

```python
y = x ** 2
plt.plot(x,y, 'r')
```

# Change the size of the plot

```python
y = x ** 2
plt.figure(figsize = (10,8))  # Here 10 is the width and 8 is the height
plt.plot(x,y, 'r')
```

# Increase the width of the line 

```python
y = x ** 2
plt.figure(figsize = (10,8))
plt.plot(x,y, 'r', linewidth = 3)
```

# Label x-axis and y-axis

```python
y = x ** 2
plt.figure(figsize = (10,8))
plt.plot(x,y, 'r', linewidth = 3)
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
```

# Making plot more informative by legend and title

```python
y = x ** 2
plt.figure(figsize = (10,8))
plt.plot(x,y, 'r', linewidth = 3, label = 'y = x ** 2')
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.legend(fontsize = 20)
plt.title('Visualization of y = x ** 2', fontsize = 25)
plt.show()
```

# Setting limits for x-axis and y-axis

```python
t = np.linspace(0,10,100)
A = 5
f = 50
sine_wave = A * np.sin(2 * np.pi * f * t)
plt.figure(figsize = (16,8))
plt.plot(t,sine_wave,linewidth = 3, label = 'SineWave')
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.ylim(-5,5) # setting limit on y-axis
plt.xlim(0,10) # setting limit on x-axis

plt.legend(fontsize = 10)
plt.title('Visualization of Sine wave', fontsize = 25)
plt.show()
```

# Using xticks and yticks

```python
t = np.linspace(0,10,100)
A = 5
f = 50
sine_wave = A * np.sin(2 * np.pi * f * t)
plt.figure(figsize = (16,8))
plt.plot(t,sine_wave,linewidth = 3, label = 'SineWave')
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.ylim(-5,5) # setting limit on y-axis
plt.xlim(0,10) # setting limit on x-axis
plt.yticks([-5,5])
plt.xticks([0,10])
plt.legend(fontsize = 10)
plt.title('Visualization of Sine wave', fontsize = 25)
plt.show()
```

# Removing axes from the plot

```python
t = np.linspace(0,10,100)
A = 5
f = 50
sine_wave = A * np.sin(2 * np.pi * f * t)
plt.figure(figsize = (16,8))
plt.plot(t,sine_wave,linewidth = 3, label = 'SineWave')
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.ylim(-5,5) # setting limit on y-axis
plt.xlim(0,10) # setting limit on x-axis
plt.axis('off')
plt.legend(fontsize = 10)
plt.title('Visualization of Sine wave', fontsize = 25)
plt.show()
```

# Making Plot nicer and nicer

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
t = np.linspace(0,10,100)
A = 5
f = 50
sine_wave = A * np.sin(2 * np.pi * f * t)
style.use('dark_background')
plt.figure(figsize = (16,8))
plt.plot(t,sine_wave,color = 'r',ls = '--', lw = 2, marker = 'o', markersize = 10, markerfacecolor = 'g',markeredgecolor = 'y',label = 'SineWave')
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.ylim(-5,5) 
plt.xlim(0,10) 
plt.grid()
plt.legend(fontsize = 10)
plt.title('Visualization of Sine wave', fontsize = 25)
plt.show()
```
